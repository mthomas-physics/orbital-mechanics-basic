```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Constants
G = 6.67430e-11 # Gravitational constant (m^3 kg^-1 s^-2)
M_sun = 1.989e30 # Mass of the Sun (kg)
M_earth = 5.972e24 # Mass of the Earth (kg)


AU = 1.496e11 # Astronomical Unit (m)
```

```python
#Place Earth at (1 AU, 0) and initial speed for circular orbit
x_earth = AU
y_earth = 0

vx_earth = 0
vy_earth =np.sqrt(G * M_sun /AU) # Orbital velocity of Earth (m/s)
```

```python
#Creatte lists to store positions
x_positions = []
y_positions = []
```

```python
dt = 3600 # Time step (1 hour in seconds)
num_steps = 365 * 24 # Simulate for one year (days * hours per day)

```

```python
for step in range(num_steps):
    # Calculate distance to the Sun
    r = np.sqrt(x_earth**2 + y_earth**2)

    # Calculate component accelerations
    ax= -G * M_sun * x_earth / r**3
    ay= -G * M_sun * y_earth / r**3

    # Update velocity
    vx_earth = vx_earth + ax * dt
    vy_earth = vy_earth + ay * dt

    # Update position
    x_earth = x_earth + vx_earth * dt
    y_earth = y_earth + vy_earth * dt

    #Store positions
    x_positions.append(x_earth)
    y_positions.append(y_earth)

    #Test energy conservation
    #kinetic_energy = 0.5 * M_earth * (vx_earth**2 + vy_earth**2)
    #potential_energy = -G * M_sun * M_earth / r
    #total_energy = kinetic_energy + potential_energy
    #print(total_energy)


```

```python
#Ploting the orbit
plt.plot(x_positions, y_positions, color ='blue')
plt.scatter(0, 0, color='yellow')
plt.xlabel("x positions(m)")
plt.ylabel("y positions(m)")
plt.title("Orbit of Earth Around the Sun")
#Equalize the axis
plt.axis('equal')
plt.legend(["Earth's Orbit", "Sun"], loc='upper right')
plt.show()
```
