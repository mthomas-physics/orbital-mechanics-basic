Invesigating the Effect of Initial Velocity on the Orbital Motion of Earth

Introduction:
During this experiment I plan to test out different initial velocities in order to determine if the velocities will result in elliptical, circular, or unbound orbits. This will be accomplished using my obrit of the earth aronud the sun simulation.


Research Question:
How does the initial velocity of the earth affect its orbit around the sun?

Predictions:
Prediction Calculations

G = 6.67430e-11 (m^3 kg^-1 s^-2)
M_sun = 1.989e30 (kg)
AU = 1.496e11 (m)

Circular Velocity = sqrt(G * M_sun/AU)
Circular Velocity = 29861.5 m/s

Escape Velocity = sqrt(2 * G * M_sun/AU)
Escape Velocity = 42230.6 m/s


| Initial Velocity(m/s) | Prediction                      |
| 15000                 | Earth falls into the sun        |
| 25000                 | Elliptical orbit                |
| 29862                 | Circular orbit                  |
| 35000                 | Elliptical orbit                |
| 42231                 | Escape tradjectory (parabolic)  |
| 46000                 | Unbound orbit (hyperbolic)      |

```python
#Simulation Setup

#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

#Constants
G = 6.67430e-11 # Gravitational constant (m^3 kg^-1 s^-2)
M_sun = 1.989e30 # Mass of the Sun (kg)
M_earth = 5.972e24 # Mass of the Earth (kg)
AU = 1.496e11 # Astronomical Unit (m)


#Place Earth at (1 AU, 0) and initial speed 
x_earth = AU
y_earth = 0

vx_earth = 0
vy_earth = [15000, 25000, 29862, 35000, 42231, 45000] # Initial velocities to test (m/s)


dt = 3600 # Time step (1 hour in seconds)
num_steps = 3 * 365 * 24 # Simulate for three years to account for non-circular orbits (3 * days * hours per day)

for v in vy_earth:
    #Reset postions for each velocity
    x = x_earth
    y = y_earth
    vx = vx_earth
    vy = v

    #Lists to store positions for plotting
    x_positions = []
    y_positions = []

    for step in range(num_steps):
        #Calculate distance from the Sun
        r = np.sqrt(x**2 + y**2)

        #Calculate Acceleration due to gravity
        ax = -G * M_sun * x / r**3
        ay = -G * M_sun * y / r**3

        #Update velocity
        vx = vx + ax * dt
        vy = vy + ay * dt

        #update positions
        x = x + vx * dt
        y = y + vy * dt

        #Update positions list
        x_positions.append(x)
        y_positions.append(y)

    #Ploting the orbit
    plt.plot(x_positions, y_positions, color ='blue')
    plt.scatter(0, 0, color='yellow')
    plt.xlabel("x positions(m)")
    plt.ylabel("y positions(m)")
    plt.title("Orbit of Earth Around the Sun at " + str(v) + "m/s")
    
    #Equalize the axis
    plt.axis('equal')
    plt.legend(["Earth's Orbit", "Sun"], loc='upper right')
    plt.show()

    #Energy Calculations
    kinetic_energy = 0.5 * M_earth* (vx**2 + vy**2)
    potential_energy = -G * M_sun * M_earth / (x**2 + y**2)**0.5
    total_energy = kinetic_energy + potential_energy
    print(f"Initial Velocity: {v} m/s, totatl_energy: {total_energy} J")




    
```
